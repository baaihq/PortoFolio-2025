```python
from google.colab import files
import pandas as pd
```

```python
# Memanggil file CSV dari github.
url = 'https://raw.githubusercontent.com/baaihq/PortoFolio-2025/main/Data-SNBP2025.csv'
df = pd.read_csv(url)
```

```python
# Memanggil DataFrame df
df
```

```python
# Mengecek tipe data dari DataFrame
df.dtypes
```

```python
# Merubah tipe data kolom ke-4 dan ke-5 menjadi integer.
df['DAYA TAMPUNG 2025'] = pd.to_numeric(df['DAYA TAMPUNG 2025'], errors='coerce').astype('Int64')
df['PEMINAT 2024'] = pd.to_numeric(df['PEMINAT 2024'], errors='coerce').astype('Int64')
```

```python
# Mengecek kembali tipe data dari DataFrame
df.dtypes
```

```python
# Hitung Rasio Keketatan (Daya Tampung / Peminat).
df['Rasio Keketatan'] = df.apply(lambda row: row.iloc[4] / row.iloc[5] if pd.notna(row.iloc[5]) and row.iloc[5] != 0 else 0, axis=1) #Jika ada pembagi NOL atau NaN, maka nilai akan dituliskan menjadi NOL.
```

```python
# Hitung Presentase (%) dari Rasio Keketatan.
df['Presentase (%)'] = (df['Rasio Keketatan'] * 100)
```

```python
# Melihat tampilan baru dari DataFrame.
df
```

```python
# Simpan DataFrame ke file CSV
df.to_csv(r'DATA-SNBP2025-OK.csv', index = False)
```
